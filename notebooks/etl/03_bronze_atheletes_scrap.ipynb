```python
# import libs
import pandas as pd
from datetime import date
import pandas_gbq
import numpy as np
import requests
from bs4 import BeautifulSoup
```

```python
athlete_name_list = []
athlete_profile_url_list = []

year_list = range(2010, 2024, 1)
for year in year_list:
# create list of surfers that will be analyzed
  year_str = str(year)
  url_athletes = 'https://www.worldsurfleague.com/athletes/tour/mct?year=' + year_str
  r = requests.get(url_athletes).text
  soup = BeautifulSoup(r, 'html.parser')
  # get html with all events of a year
  all_athletes_html = soup.find_all('a', attrs = {'class':'athlete-name'})
  for i in range(len(all_athletes_html)):
    athlete_name = all_athletes_html[i].text
    athlete_profile_url = all_athletes_html[i]['href']
    # append values
    athlete_name_list.append(athlete_name)
    athlete_profile_url_list.append(athlete_profile_url)
  print('Atheletes from ', year, ' ranking has been collected.')

# drop duplicates from lists
athlete_name_list = list(dict.fromkeys(athlete_name_list))
athlete_profile_url_list = list(dict.fromkeys(athlete_profile_url_list))
```

```python
# create empty lists that will be filled with atheletes informations
country_list = []
stance_list = []
first_season_list = []
birth_date_list = []
height_list = []
weight_list = []

#
for athlete_url in athlete_profile_url_list:
  r = requests.get(athlete_url).text
  soup = BeautifulSoup(r, 'html.parser')
  athlete_html = soup.find_all('div', attrs = {'class':'new-athlete-bio-stats'})
  athlete_html = athlete_html[0]
  # set empty variables, some atheletes don't have all informations
  stance = ''
  first_season = ''
  age = ''
  height = ''
  weight = ''
  country = ''
  # diferent from others, country is outside athlete_html
  try:
    country = soup.find_all('div', attrs = {'class':'country-name'})[0].text
  except:
    pass
  #
  for i in range(len(athlete_html.find_all('div', attrs = {'class':'value'}))):
    # stance
    if 'Stance' in athlete_html.find_all('li')[i].text:
      stance = athlete_html.find_all('li')[i].find('div', {'class':'value'}).text
    # first season
    if 'First season' in athlete_html.find_all('li')[i].text:
      first_season = athlete_html.find_all('li')[i].find('div', {'class':'value'}).text
    # age
    if 'Age' in athlete_html.find_all('li')[i].text:
      birth_date = athlete_html.find_all('li')[i].find('span', attrs = {'class':'metric'}).text
      #age = re.sub("[^0-9]", "", birth_date)
    # height
    if 'Height' in athlete_html.find_all('li')[i].text:
      height = athlete_html.find_all('div', attrs = {'class':'value'})[i].find('span', attrs = {'class':'metric'}).text
      #height = re.sub("[^0-9]", "", height)
    # weight
    if 'Weight' in athlete_html.find_all('li')[i].text:
      weight = athlete_html.find_all('div', attrs = {'class':'value'})[i].find('span', attrs = {'class':'metric'}).text
      #weight = re.sub("[^0-9]", "", weight)
    # home_town
    #if 'Hometown' in athlete_html.find_all('li')[i].text:
    #  home_town = athlete_html.find_all('div', attrs = {'class':'value'})[i].text
    else:
      pass
  print('Athelete profile information from ', athlete_url, ' has been collected.')
  # append values
  stance_list.append(stance)
  first_season_list.append(first_season)
  birth_date_list.append(birth_date)
  height_list.append(height)
  weight_list.append(weight)
  #home_town_list.append(home_town)
  country_list.append(country)
```

```python
# create df
df_athletes = pd.DataFrame(
                        {
                         'athlete_name': athlete_name_list,
                         'athlete_country':country_list,
                         'athlete_profile_url': athlete_profile_url_list,
                         'athlete_stance': stance_list,
                         'athlete_first_season': first_season_list,
                         'athlete_birth_date': birth_date_list,
                         'athlete_height': height_list,
                         'athlete_weight': weight_list
                        })
```
